## Brief Introduction to Software Testing Concepts

The textbook definition of software testing states that it's the process of executing a program or application to find any bugs. For medium to large-scale projects, software testing is done to determine if the software behaves as intended under various sets of inputs and conditions.


## Unit Testing

Unit testing is a software testing method in which individual components of the program, called `units`, are tested independently with all the required conditions.


## Test Automation

Test automation is the automated execution and reporting of the outcome of test scenarios and cases. Various tools like code libraries and third-party APIs are used for unit testing.

# Brief Introduction to doctest
---

**doctest** is the lightweight unit-testing framework in Python that uses docstrings to test automation. It is packaged with the Python interpreter.

The code listed below is a simple example of a test module with two functions and two tests for each function

```python
"""
Sample doctest test module - test_module02
"""

def mul(a, b):
  """
  >>> mul(2, 3)
  6

  >>> mul('a', 2)
  'aa'
  """

  return a * b


def add(a, b):
  """
  >>> add(2, 3)
  5

  >>> add('a', 'b')
  'ab'
  """

  return a + b
```

The test cases are mentioned as docstrings for the modules. There is nothing specifically calling the doctest in the code itself.

In order to see doctest in action, the following command needs to be entered:

`python3 -m doctest -v test_module02.py`

The following output is returned:

Screenshot 2026-01-07 at 15.45.16.png

**doctest** works by parsing docstrings. Whenever doctest finds an interactive Python prompt in the documentation of a module, it treats its output as the expected output. Then it runs the module and its members by referring to the docstrings. It compares the actual output against the output specified in the docstrings. Then it marks the test pass or fail.

* `-m doctest` is used while executing the module to let the interpreter know that the doctest module needs to be used while executing the code.

* `-v` stands for **verbose** mode. Without it, the test will not produce any output unless it fails. Using verbose produces an execution log irrespective of whether the test passes or fails.

Tests can fail due to one or more of the following reasons:

* faulty logic in the code
* faulty input into the code
* faulty test case

# Pydoc

Just like **doctest**, there is another useful utility to view the documentation of a module. It comes preinstalled with Python and is known as **Pydoc**.
